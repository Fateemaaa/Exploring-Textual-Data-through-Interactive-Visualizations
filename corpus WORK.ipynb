```python
import json
from nltk.tokenize import word_tokenize

# Load the dataset
with open('corpus.json', 'r') as json_file:
    corpus_data = json.load(json_file)

# Tokenize the 'name' field in each entry
for entry in corpus_data:
    if isinstance(entry, dict):  # Check if the entry is a dictionary
        movie_name = entry.get('name', '')
        tokens = word_tokenize(movie_name)

        # Print the tokens for each movie
        print(f"Movie: {movie_name}")
        print(f"Tokens: {tokens}")
        print("------")

# If you want to save the tokenized data back to a new file
with open('tokenized_dataset.json', 'w') as tokenized_file:
    json.dump(corpus_data, tokenized_file, indent=2)
    
    
with open('tokenized_dataset.json', 'r') as tokenized_file:
    print(tokenized_file.read())
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

# Load the cleaned dataset
with open('cleaned_corpus.json', 'r') as cleaned_file:
    cleaned_data = json.load(cleaned_file)

# Convert the cleaned data to a DataFrame
df_cleaned = pd.DataFrame(cleaned_data)

# Display basic information about the DataFrame
print("Basic Information about the Cleaned DataFrame:")
print(df_cleaned.info())

# Display summary statistics of the numerical columns
print("\nSummary Statistics of the Numerical Columns:")
print(df_cleaned.describe())

# Display the first few rows of the cleaned DataFrame
print("\nFirst few rows of the Cleaned DataFrame:")
print(df_cleaned.head())

# Tokenize the 'name' field in each entry
df_cleaned['tokens'] = df_cleaned['name'].apply(word_tokenize)

# Display the tokenized DataFrame
print("\nTokenized DataFrame:")
print(df_cleaned[['name', 'tokens']].head())


```

```python
# Visualize the distribution of token lengths
df_cleaned['token_length'] = df_cleaned['tokens'].apply(len)

# Plot histogram of token lengths
plt.figure(figsize=(12, 6))
sns.histplot(df_cleaned['token_length'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Token Lengths')
plt.xlabel('Token Length')
plt.ylabel('Frequency')
plt.show()

```

```python
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from wordcloud import WordCloud
# Generate a word cloud to visualize the most common words in movie names
all_tokens = [token for sublist in df_cleaned['tokens'] for token in sublist]
wordcloud = WordCloud(width=400, height=200, background_color='black').generate(' '.join(all_tokens))

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Movie Names')
plt.show()
```

```python
# Distribution of the number of tokens per movie
df_cleaned['token_length'] = df_cleaned['tokens'].apply(len)
plt.figure(figsize=(8, 4))
sns.barplot(x=df_cleaned['token_length'].value_counts().index, y=df_cleaned['token_length'].value_counts().values, color='lightcoral')
plt.title('Distribution of Number of Tokens per Movie')
plt.xlabel('Number of Tokens')
plt.ylabel('Number of Movies')
plt.show()
```
